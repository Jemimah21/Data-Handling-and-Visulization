```python
import pandas as pd
import random
```

```python
Age_Group=["<21","21-24","24-27","27-30","30-33","33-36","36-39","39-42","42-45","45-48","48-51","51-54","54-57","57-60",">60"]
Total_no_of_loans=[random.randint(10,9000) for i in range(15)]
no_of_bad_loans=[random.randint(10,200) for i in range(15)]
no_of_good_loans=[Total_no_of_loans[i]-no_of_bad_loans[i] for i in range(15)]
print(no_of_good_loans)
```

```python
Percentage_of_bad_loans=[((no_of_bad_loans[i]/Total_no_of_loans[i])*100) for i in range(15)]
Percentage_of_good_loans=[((no_of_good_loans[i]/Total_no_of_loans[i])*100) for i in range(15)]
```

```python
data={
    'Age':Age_Group,
    'Total Loans':Total_no_of_loans,
    'Bad Loans':no_of_bad_loans,
    'Good Loans':no_of_good_loans,
    'Percentage Bad Loans':Percentage_of_bad_loans,
    'Percentage Good Loans':Percentage_of_good_loans

}
df = pd.DataFrame(data)

print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
Age_Group = ['<20', '20-30', '30-40', '40-50']
Total_no_of_loans = [100, 200, 300, 400]
no_of_bad_loans = [10, 20, 30, 40]
no_of_good_loans = [90, 180, 270, 360]
Percentage_of_bad_loans = [10, 10, 10, 10]
Percentage_of_good_loans = [90, 90, 90, 90]

data = {
    'Age': Age_Group,
    'Total Loans': Total_no_of_loans,
    'Bad Loans': no_of_bad_loans,
    'Good Loans': no_of_good_loans,
    'Percentage Bad Loans': Percentage_of_bad_loans,
    'Percentage Good Loans': Percentage_of_good_loans
}

df = pd.DataFrame(data)
fig, axes = plt.subplots(2, 3, figsize=(10, 6))
for i, ax in enumerate(axes.flat):
    if i == 0:
        df.plot(x='Age', y='Total Loans', kind='bar', ax=ax, color='skyblue')
        ax.set_title('Total Loans')
    elif i == 1:
        df.plot(x='Age', y='Bad Loans', kind='bar', ax=ax, color='red')
        ax.set_title('Bad Loans')
    elif i == 2:
        df.plot(x='Age', y='Good Loans', kind='bar', ax=ax, color='green')
        ax.set_title('Good Loans')
    elif i == 3:
        df.plot(x='Age', y='Percentage Bad Loans', kind='bar', ax=ax, color='orange')
        ax.set_title('Percentage Bad Loans')
    elif i == 4:
        df.plot(x='Age', y='Percentage Good Loans', kind='bar', ax=ax, color='purple')
        ax.set_title('Percentage Good Loans')
    else:
        ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
Age_Group = ['<20', '20-30', '30-40', '40-50']
Total_no_of_loans = [100, 200, 300, 400]
no_of_bad_loans = [10, 20, 30, 40]
no_of_good_loans = [90, 180, 270, 360]
Percentage_of_bad_loans = [10, 10, 10, 10]
Percentage_of_good_loans = [90, 90, 90, 90]

data = {
    'Age': Age_Group,
    'Total Loans': Total_no_of_loans,
    'Bad Loans': no_of_bad_loans,
    'Good Loans': no_of_good_loans,
    'Percentage Bad Loans': Percentage_of_bad_loans,
    'Percentage Good Loans': Percentage_of_good_loans
}

df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))
plt.subplot(2, 3, 1)
sns.barplot(x='Age', y='Total Loans', data=df, palette='Blues')
plt.title('Total Loans')

plt.subplot(2, 3, 2)
sns.barplot(x='Age', y='Bad Loans', data=df, palette='Reds')
plt.title('Bad Loans')

plt.subplot(2, 3, 3)
sns.barplot(x='Age', y='Good Loans', data=df, palette='Greens')
plt.title('Good Loans')
plt.subplot(2, 3, 4)
plt.pie(df['Percentage Bad Loans'], labels=df['Age'], autopct='%1.1f%%', colors=sns.color_palette('Oranges'))
plt.title('Percentage of Bad Loans')

plt.subplot(2, 3, 5)
plt.pie(df['Percentage Good Loans'], labels=df['Age'], autopct='%1.1f%%', colors=sns.color_palette('Purples'))
plt.title('Percentage of Good Loans')

plt.subplot(2, 3, 6)
plt.plot(df['Age'], df['Total Loans'], marker='o', color='skyblue', label='Total Loans')
plt.plot(df['Age'], df['Bad Loans'], marker='o', color='red', label='Bad Loans')
plt.plot(df['Age'], df['Good Loans'], marker='o', color='green', label='Good Loans')
plt.legend()
plt.title('Loan Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Loans')

plt.tight_layout()
plt.show()
```
