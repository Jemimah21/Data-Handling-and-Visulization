**Make an array of student marks of a course(generate random marks).Assume fail criteria is less than 30. find number of students passed and filter failed marks**

```python
std_marks = [random.randint(0, 100) for _ in range(50)]
passed_stds = [mark for mark in std_marks if mark >= 30]
failed_marks = [mark for mark in std_marks if mark < 30]

no_of_stds_passed= len(passed_stds)

print(f"Number of students passed: {no_of_stds_passed}")
print(f"Failed marks: {failed_marks}")
```
