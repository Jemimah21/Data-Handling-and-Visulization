```python
import numpy as np
import matplotlib.pyplot as plt
data =np.random.random((13,16))
plt.imshow(data,cmap='magma')
plt.title("2-D heat maps")
plt.colorbar()
plt.show()
```

```python
import seaborn as sns
data=np.random.randint(low=14,high=100,size=(20,20))
hm=sns.heatmap(data=data,annot=True)
plt.show()
```

```python
import seaborn as sns
data=np.random.randint(low=14,high=100,size=(5,5))
linewidths = 2
linecolor = "yellow"
hm=sns.heatmap(data=data,annot=True,linewidths=linewidths, linecolor=linecolor)
plt.show()
```
