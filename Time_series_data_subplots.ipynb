```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data1={
    "Date1":pd.to_datetime(["2022-01-01","2022-02-21","2022-03-03","2022-04-04","2022-05-05","2022-06-06","2022-07-07","2022-08-05","2022-09-05","2022-10-05","2022-11-05","2022-12-05"]),
    "Value":[10,30,22,28,100,42,48,55,90,75,98,110]
}
data={
    "Date":pd.to_datetime(["2022-01-01","2022-02-21","2022-03-03","2022-04-04","2022-05-05","2022-06-06","2022-07-07","2022-08-05","2022-09-05","2022-10-05","2022-11-05","2022-12-05"]),
    "Value":[10,15,22,28,35,42,48,55,60,75,98,105]
}
```

```python
df=pd.DataFrame(data)
df.set_index("Date",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["Value"])
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time series with trend")
plt.grid(True)
plt.show()
```

```python
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/monthly-sunspots.csv"
data = pd.read_csv(url, parse_dates=['Month'], index_col='Month')
print(data)
```

```python
plt.figure(figsize=(7, 5))
plt.plot(data)
plt.xlabel('Date')
plt.ylabel('Number of Sunspots')
plt.title('Monthly Sunspots Line Plot')
plt.grid(True)
plt.show()
```
