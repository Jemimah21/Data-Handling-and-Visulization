```python
mymap = folium.Map(location=[12.9655881, 77.6036377], zoom_start=10)
marker_locations = [
    [12.9655881, 77.6036377],
    [12.9715981, 77.5945627],
    [12.9660917, 77.5875011]
]

marker_popups = [
    'Presidency University, Bengaluru',
    'Location 2',
    'Location 3'
]

for location, popup in zip(marker_locations, marker_popups):
    folium.Marker(location=location, popup=popup).add_to(mymap)
mymap
```

```python
mymap = folium.Map(location=[12.9655881, 77.6036377], zoom_start=10)
marker_locations = [
    [12.9655881, 77.6036377],
    [12.9715981, 77.5945627],
    [12.9660917, 77.5875011]
]

marker_popups = [
    'Presidency University, Bengaluru',
    'Location 2',
    'Location 3'
]

marker_colors = ['red', 'blue', 'green']

for location, popup, color in zip(marker_locations, marker_popups, marker_colors):
    folium.Marker(location=location, popup=popup, icon=folium.Icon(color=color)).add_to(mymap)
mymap
```
