```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={
    "Date":pd.to_datetime(["2022-01-01","2022-02-21","2022-03-03","2022-04-04","2022-05-05","2022-06-06","2022-07-07","2022-08-05","2022-09-05","2022-10-05","2022-11-05","2022-12-05"]),
    "Value":[10,15,22,28,35,42,48,55,60,75,98,105]
}
```

```python
df=pd.DataFrame(data)
df.set_index("Date",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["Value"],marker='o')
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time series with trend")
plt.grid(True)
plt.show
```
